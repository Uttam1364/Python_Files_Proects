# Netflix Stock Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from datetime import datetime
```

```python
d=pd.read_csv("NFLX.csv")
d
```

```python
d.head()
```

```python
sns.set(rc={'figure.figsize':(10,5)})
## Ye kya hai muhe bhi nahi maloom
```

```python
## Setting date as the index
d['Date']=pd.to_datetime(d['Date'])
d=d.set_index('Date')
d.head()
```

```python
## Volume verses date graph
sns.lineplot(x=d.index,y=d['Volume'],label='Volume',color='green')
plt.title("Volume v/s Time Graph")
```

```python
#Netflix stock price graph:High, Open, Close
d.plot(y=["High","Open","Close"],title="Netflix stock price")
```

# Netflix stock price graph:Day, Month, Year


```python
fig,(ax1,ax2,ax3)=plt.subplots(3,figsize=(15,10))
d.groupby(d.index.day).mean().plot(y='Volume',ax=ax1,xlabel='Day')
d.groupby(d.index.month).mean().plot(y='Volume',ax=ax2,xlabel='Month')
d.groupby(d.index.year).mean().plot(y='Volume',ax=ax3,xlabel='Year')
```

# Top 5 dates with Highest stock price

```python
a=d.sort_values(by='High',ascending=False).head(5)
a['High']
```

# Top 5 dates with Highest stock price

```python
a=d.sort_values(by='Low',ascending=True).head(5)
a['Low']
```

# Trend of High and Low stock prices

```python
fig,axes=plt.subplots(nrows=1,ncols=2,sharex=True,figsize=(12,5))
fig.suptitle('High & Low values stock per period of time',fontsize=18)
sns.lineplot(ax=axes[0],y=d['High'],x=d.index,color='green')
sns.lineplot(ax=axes[1],y=d['Low'],x=d.index,color='red')
```
